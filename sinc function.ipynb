```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a sinc signal
f = 5 # Hz
sample_rate = 50 # Hz
interval = 10 # seconds

t = np.arange(-interval/2, interval/2, 1/sample_rate)
signal = np.sin(np.pi*t*f)/(t*np.pi)
plt.plot(t, signal)
plt.xlabel('x')
plt.ylabel('amplitude');
```

```python
N = 1024;
w = (np.fft.fft(signal, N));
freq = np.fft.fftfreq(N, d=1.0/sample_rate)
plt.plot(freq, np.abs(w));
plt.xlabel('frequency (Hz)');
plt.ylabel('magnitude');
plt.xlim(-3,3)
```
